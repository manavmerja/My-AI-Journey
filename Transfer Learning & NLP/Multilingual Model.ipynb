```python
from transformers import pipeline

# 1. Pipeline with a Specific Multilingual Model
# पहले हम default English model यूज़ कर रहे थे।
# अब हम 'cardiffnlp/twitter-xlm-roberta-base-sentiment' बुला रहे हैं।
classifier = pipeline("sentiment-analysis", model="cardiffnlp/twitter-xlm-roberta-base-sentiment")

# 2. Test with Hinglish
text = "Ye movie bekar thi"
result = classifier(text)

print(f"Sentence: {text}")
print(f"Result: {result}")

# 3. Test with Pure Hindi
text2 = "यह फिल्म बेहतरीन थी"
result2 = classifier(text2)

print(f"\nSentence: {text2}")
print(f"Result: {result2}")
```
