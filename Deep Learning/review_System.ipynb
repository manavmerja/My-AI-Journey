```python
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences

# 1. Hamara Raw Data (Tokens)
# Review A (Chota), Review B (Bada)
reviews = [
    [10, 20],               # "Amazing movie"
    [5, 20, 30, 40, 50, 60] # "Worst movie ever I slept..."
]

# 2. Magic Command: pad_sequences
# maxlen=5: Humein har review ki length 5 chahiye
# padding='post': Zeros 'peeche' lagao (10, 20, 0, 0...)
# truncating='post': Extra words 'peeche' se kato
padded_reviews = pad_sequences(reviews, maxlen=5, padding='pre', truncating='pre')

print("Processed Reviews (Ready for RNN):")
print(padded_reviews)
```
