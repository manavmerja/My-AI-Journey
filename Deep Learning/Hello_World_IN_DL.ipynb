```python
import tensorflow as tf

# 1. Data Load karna (MNIST Dataset)
# Ye dataset TensorFlow mein pehle se hota hai
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Preprocessing (Normalization)
# Pixel value 0 se 255 hoti hai. Hum use 0 se 1 ke beech la rahe hain taaki calculation fast ho.
x_train, x_test = x_train / 255.0, x_test / 255.0

# 3. Model Banana (The Brain Structure) ğŸ§ 
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),  # Input Layer: 2D image ko 1D line mein kholna
  tf.keras.layers.Dense(128, activation='relu'),  # Hidden Layer: 128 Neurons (Sochne ke liye)
  tf.keras.layers.Dense(10, activation='softmax') # Output Layer: 10 Neurons (0 se 9 tak digits)
])

# 4. Compile (Teacher assign karna)
# Optimizer: Adam (Jo weights update karega)
# Loss: Sparse Categorical Crossentropy (Galti napne ke liye)
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 5. Train (Padhai shuru) ğŸ“š
# Epochs=5 matlab pura data 5 baar dekho
model.fit(x_train, y_train, epochs=5)

# 6. Evaluation (Test)
print("\nEvaluation on Test Data:")
model.evaluate(x_test, y_test)
```

```python
import tensorflow as tf

# 1. Data Load & Normalize (Wahi purana step)
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape: CNN ko batana padta hai ki color channel kitne hain (28, 28, 1)
# 1 = Grayscale (Black & White). Agar rangeen hota to 3 hota (RGB).
x_train = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test = x_test.reshape(-1, 28, 28, 1) / 255.0

# 2. CNN Model Banana ğŸ—ï¸
model = tf.keras.models.Sequential([
    # Layer 1: Convolution (Scanner)
    # 32 Filters (32 alag-alag patterns dhundega)
    # (3, 3) ka kernel size (Torch ka size)
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),

    # Layer 2: Max Pooling (Compressor)
    # 2x2 ka box lekar sabse bada number uthayega (Image size aadha ho jayega)
    tf.keras.layers.MaxPooling2D((2, 2)),

    # Layer 3: Flatten (Ab hum 2D se 1D mein ayenge decision lene ke liye)
    tf.keras.layers.Flatten(),

    # Layer 4: Dense (Sochne wala dimag)
    tf.keras.layers.Dense(64, activation='relu'),

    # Layer 5: Output (0-9 Digits)
    tf.keras.layers.Dense(10, activation='softmax')
])

# 3. Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 4. Train (Sirf 5 Epochs)
model.fit(x_train, y_train, epochs=5)

# 5. Evaluate
print("\nFinal Test Accuracy:")
model.evaluate(x_test, y_test)
```

```python
# --- Step 1: Save the Model ---
# Hum iska naam 'digit_recognizer.keras' rakh rahe hain
# .keras naya aur recommended format hai
model.save('digit_recognizer.keras')
print("âœ… Model Saved Successfully!")

# --- Step 2: Load the Model (Proof) ---
# Maan lijiye humne PC restart kiya, ab wapas load kar rahe hain
loaded_model = tf.keras.models.load_model('digit_recognizer.keras')

print("\nChecking Loaded Model Accuracy:")
# Dekhte hain kya ye wahi purana model hai?
loaded_model.evaluate(x_test, y_test)
```
