```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

# --- Step 1: Data Load Karna ---
# num_words=10000: Hum sirf top 10,000 sabse common words rakhenge (Baaki rare words bekar hain)
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)

# --- Step 2: Padding (Pre-padding logic) ---
# Har review ko exactly 100 words ka banayenge
x_train = pad_sequences(x_train, maxlen=100, padding='pre', truncating='pre')
x_test = pad_sequences(x_test, maxlen=100, padding='pre', truncating='pre')

# --- Step 3: RNN Model Banana ğŸ§  ---
model = tf.keras.models.Sequential([
    # Layer 1: Embedding (Sabdo ko vector banana)
    # 10000 words vocabulary, har word ko 32 numbers (vector) mein badlo
    tf.keras.layers.Embedding(input_dim=10000, output_dim=32, input_length=100),

    # Layer 2: Simple RNN (Loop wala dimag)
    # 32 Neurons jo sequence yaad rakhenge
    tf.keras.layers.SimpleRNN(32),

    # Layer 3: Output (Hit ya Flop?)
    # Sigmoid kyun? Kyunki humein 0 (Flop) ya 1 (Hit) chahiye
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# --- Step 4: Compile & Train ---
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("Training Start... (Yeh thoda time lega)")
model.fit(x_train, y_train, epochs=5, batch_size=128)

# --- Step 5: Test ---
print("\nFinal Accuracy on Test Data:")
model.evaluate(x_test, y_test)
```
