```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

# --- Step 1: Data Load & Preprocessing (Same as before) ---
print("Loading Data...")
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)

# Padding (Taaki sab barabar ho)
x_train = pad_sequences(x_train, maxlen=100, padding='pre', truncating='pre')
x_test = pad_sequences(x_test, maxlen=100, padding='pre', truncating='pre')

# --- Step 2: The LSTM Model ğŸ§  ---
model = tf.keras.models.Sequential([
    # Layer 1: Embedding (Words to Vectors)
    tf.keras.layers.Embedding(input_dim=10000, output_dim=32),

    # Layer 2: LSTM (The Upgrade!) ğŸš€
    # Humne SimpleRNN hata kar LSTM laga diya
    # 64 units ka matlab: Iske paas 64 parallel diaries hain yaad rakhne ke liye
    tf.keras.layers.LSTM(64),

    # Layer 3: Output (Hit or Flop)
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# --- Step 3: Compile & Train ---
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("Training LSTM... (Sabr rakhein, ye thoda heavy hai)")
model.fit(x_train, y_train, epochs=5, batch_size=128)

# --- Step 4: Final Test ---
print("\nLSTM Evaluation:")
model.evaluate(x_test, y_test)
```
