```python
import matplotlib.pyplot as plt

# 1. Data (Comparison)
years = [2010, 2011, 2012, 2013, 2014, 2015]
sachin_runs = [1500, 1000, 800, 200, 0, 0]   # Sachin ka career end ho raha hai
kohli_runs = [500, 900, 1200, 1500, 1800, 2100] # Kohli ka career rise ho raha hai

# 2. Plotting (Multiple Lines)
# Sachin: Red Color, Dashed Line (--), Circle Marker (o)
plt.plot(years, sachin_runs, color='red', linestyle='--', marker='o', label="Sachin")

# Kohli: Blue Color, Solid Line (-), Triangle Marker (^)
plt.plot(years, kohli_runs, color='blue', linestyle='-', marker='^', linewidth=2, label="Kohli")

# 3. Decoration
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Career Comparison: Sachin vs Kohli")
plt.grid(True)  # Grid (Jali) on kar di taaki points padhna aasaan ho
plt.legend()    # Ye batata hai ki Red kaun hai aur Blue kaun hai

# 4. Display
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Data
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

# 2. Plotting (Bar Chart)
# color='skyblue' se bars ka rang set kiya, width=1.5 se bars ki motai set ki
plt.bar(years, runs, color='skyblue', width=1.5)

# 3. Decoration
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs (Bar Chart)")

# 4. Show
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Data (Virat Kohli's Approximate Runs)
formats = ["Test", "ODI", "T20"]
runs = [8000, 13000, 4000]

# 2. Styling (Thoda professional banate hain)
colors = ["gold", "skyblue", "lightgreen"] # Alag rang
explode = [0, 0.1, 0]  # ODI wale hisse ko thoda bahar nikalenge (Highlight karne ke liye)

# 3. Plotting
# autopct='%1.1f%%' -> Ye percentage calculate karke graph pe likhta hai
plt.pie(runs, labels=formats, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True)

# 4. Decoration
plt.title("Virat Kohli's Run Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Data (Padhai ke ghante vs Marks)
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [20, 30, 40, 50, 60, 70, 80, 90]

# 2. Plotting (Scatter)
# color='green' -> Hara rang
# s=100 -> Size of dots (Points ko bada dikhane ke liye)
# marker='*' -> Points ko Star shape diya
plt.scatter(study_hours, exam_scores, color='green', s=100, marker='*')

# 3. Decoration
plt.title("Study Hours vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.grid(True)

# 4. Show
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Data (7 Days ka schedule)
days = [1, 2, 3, 4, 5, 6, 7]

# Har din kitne ghante kahan gaye
sleeping = [7, 8, 6, 11, 7, 8, 9]
studying = [2, 3, 4, 3, 2, 6, 4]
playing  = [7, 8, 7, 2, 2, 5, 3]
eating   = [8, 5, 7, 8, 13, 5, 8]  # Bacha hua time

# 2. Plotting (Stackplot)
# Colors: m=magenta, c=cyan, r=red, k=black
plt.stackplot(days, sleeping, studying, playing, eating, 
              labels=['Sleeping', 'Studying', 'Playing', 'Eating'], 
              colors=['m', 'c', 'r', 'k'])

# 3. Decoration
plt.xlabel('Days')
plt.ylabel('Hours')
plt.title('Weekly Routine Breakdown')
plt.legend(loc='upper left') # Legend ko kone mein rakha taaki graph na chupe

# 4. Show
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Fake Data (YouTube Viewers ki Age)
# Ye bohot saare random ages hain (10 se 60 saal ke beech)
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55, 10, 15, 60, 22, 22, 25, 45]

# 2. Plotting (Histogram)
# bins=[10, 20, 30, 40, 50, 60] -> Humne khud ranges batayi hain
# edgecolor='black' -> Bars ke beech line dikhegi
plt.hist(ages, bins=[10, 20, 30, 40, 50, 60], edgecolor='black', color='orange')

# 3. Decoration
plt.title("Age Distribution of Viewers")
plt.xlabel("Age Groups")
plt.ylabel("Number of Viewers")

# 4. Show
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Data
x = np.arange(1, 11) # 1 se 10 tak numbers
y1 = x * 2           # Double
y2 = x ** 2          # Square
y3 = np.sqrt(x)      # Square Root
y4 = 10 - x          # Reverse

# 2. Figure Size set karte hain (Thoda bada canvas)
plt.figure(figsize=(10, 8))

# --- Graph 1 (Top Left) ---
plt.subplot(2, 2, 1)  # 2 rows, 2 cols, 1st position
plt.plot(x, y1, 'r--') # Red Dashed Line
plt.title("Double (x * 2)")

# --- Graph 2 (Top Right) ---
plt.subplot(2, 2, 2)  # 2nd position
plt.plot(x, y2, 'g*-') # Green Star Line
plt.title("Square (x^2)")

# --- Graph 3 (Bottom Left) ---
plt.subplot(2, 2, 3)  # 3rd position
plt.plot(x, y3, 'b^') # Blue Triangles (Scatter jaisa)
plt.title("Square Root")

# --- Graph 4 (Bottom Right) ---
plt.subplot(2, 2, 4)  # 4th position
plt.plot(x, y4, 'k-') # Black Line
plt.title("Reverse (10 - x)")

# 3. Final Touch
plt.tight_layout() # Ye graphs ko chipakne se rokta hai (Clean look)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Setup Theme (Graph sunder dikhega)
sns.set_theme(style="darkgrid")  # Options: whitegrid, dark, white, ticks 

# 2. Built-in Data Load karte hain ('tips' dataset - Hotel Bill ka data)
# Ye Seaborn ke paas pehle se hota hai practice ke liye [cite: 623]
tips = sns.load_dataset("tips")

print(tips.head()) # Data dekhlo kaisa hai

# 3. Plotting (Scatter)
# x: Bill Amount, y: Tip Amount
# hue="time": Lunch aur Dinner ko alag rang dega apne aap! 
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time", s=100)

# 4. Decoration
plt.title("Total Bill vs Tip (Colored by Time)")
plt.show()
```
