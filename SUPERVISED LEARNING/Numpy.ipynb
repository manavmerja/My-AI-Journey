```python
# Ek 2x3 matrix (2 rows, 3 columns)
data = np.array([
    [10, 20, 30],
    [5,  5,  5]
])

# Task: Har column ka maximum number nikalo (Axis 0)
max_vals = np.max(data, axis=0)
print(max_vals)
```

```python
data = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

# Column ka Mean (Average) nikalo
mean = data.mean(axis=0) 
print("Mean:", mean)

# Pure data me se mean ko minus karo (Broadcasting)
centered_data = data - mean
print("Centered Data:\n", centered_data)
```
