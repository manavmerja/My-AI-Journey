```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # 1. Model Import kiya

# --- Step 1: Data Taiyaar Karna ---
# X (Feature) hamesha 2D array hona chahiye (isliye reshape kiya)
experience = np.array([1, 2, 3, 4, 5]).reshape(-1, 1) 
salary = np.array([20000, 30000, 40000, 50000, 60000]) # y (Target)

# --- Step 2: Model Banana aur Train Karna ---
model = LinearRegression()  # Model create kiya
model.fit(experience, salary) # Training shuru (Computer pattern seekh raha hai)

# --- Step 3: Magic (Prediction) ---
# Ab hum puchte hain: "Agar 3.5 saal ka experience hai, toh salary kya hogi?"
new_exp = np.array([[3.5]]) 
predicted_salary = model.predict(new_exp)

print(f"3.5 Years Experience ke liye Salary hogi: ₹{predicted_salary[0]}")

# --- Step 4: Visualization (Best Fit Line) ---
plt.scatter(experience, salary, color='blue', label='Asli Data') # Points
plt.plot(experience, model.predict(experience), color='red', label='Best Fit Line') # Line
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# --- Step 1: Data Taiyaar Karna (2 Features) ---
# X = [Experience, Interview_Score]
X = np.array([
    [1, 5],   # 1 Year Exp, 5/10 Score
    [2, 9],   # 2 Years Exp, 9/10 Score (Accha Score!)
    [3, 6],   # 3 Years Exp, 6/10 Score
    [4, 10],  # 4 Years Exp, 10/10 Score (Topper!)
    [5, 5]    # 5 Years Exp, 5/10 Score
])

y = np.array([25000, 38000, 42000, 60000, 55000]) # Asli Salary

# --- Step 2: Model Train Karna ---
model = LinearRegression()
model.fit(X, y)

# --- Step 3: Secrets Reveal (Coefficients) ---
print(f"Intercept (Base Salary): {model.intercept_:.2f}")
print(f"Weights [Experience, Interview Score]: {model.coef_}")

# --- Step 4: Prediction ---
# Socho ek banda hai: 3 Saal Experience, aur 9/10 Score
new_candidate = np.array([[3, 9]])
predicted_salary = model.predict(new_candidate)

print(f"Predicted Salary for New Candidate: {predicted_salary[0]:.2f}")
```
