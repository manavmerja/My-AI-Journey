```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
import numpy as np

# --- Step 1: Data ---
# [Salary, Credit_History]
# Salary: 1=High, 0=Low
# Credit: 1=Good, 0=Bad
X = np.array([
    [1, 1], # High Salary, Good Credit
    [1, 0], # High Salary, Bad Credit
    [0, 1], # Low Salary, Good Credit
    [0, 0], # Low Salary, Bad Credit
    [1, 1], # High Salary, Good Credit
    [0, 1]  # Low Salary, Good Credit (Socho iska kya hoga?)
])

# Loan Approved? (1=Yes, 0=No)
# Logic: Loan tabhi milega jab (Salary=High OR Credit=Good) - chalo ye logic test karte hain
y = np.array([1, 0, 0, 0, 1, 0]) 
# Maine logic thoda strict kar diya: Sirf (High Salary + Good Credit) = Yes. Baaki sab No.

# --- Step 2: Model ---
model = DecisionTreeClassifier(criterion='entropy') # Entropy use karenge
model.fit(X, y)

# --- Step 3: Visualize the Tree ---
plt.figure(figsize=(10, 6))
tree.plot_tree(model, feature_names=['Salary', 'Credit_History'], class_names=['No', 'Yes'], filled=True)
plt.title("Decision Tree Logic")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# --- Step 1: Data wapas banate hain ---
# Computer purana data bhool gaya tha, isliye dobara define kar rahe hain
X = np.array([
    [1, 1], # High Salary, Good Credit
    [1, 0], # High Salary, Bad Credit
    [0, 1], # Low Salary, Good Credit
    [0, 0], # Low Salary, Bad Credit
    [1, 1], # High Salary, Good Credit
    [0, 1]  # Low Salary, Good Credit
])

# Target: 1=Approve, 0=Reject
y = np.array([1, 0, 0, 0, 1, 0])

# --- Step 2: Model Banao (100 Trees ka Jungle) ---
model = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=42)

# --- Step 3: Train Karo ---
model.fit(X, y)

# --- Step 4: Predict Karo (Voting Time) ---
# Naya Customer: High Salary (1), Good Credit (1)
prediction = model.predict([[1, 1]])

print(f"Final Decision by Committee: {'Approve' if prediction[0] == 1 else 'Reject'}")
```
