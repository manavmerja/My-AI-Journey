```python
class Employee:
    company = "HP"
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def printDatails(self):
        print(f"The company of {self.name} is {self.company} with salary {self.salary}")

e = Employee("harry", 47000)
e.printDatails()
```
