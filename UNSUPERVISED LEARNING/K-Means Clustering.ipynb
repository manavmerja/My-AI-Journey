```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# --- Step 1: Generate Data (No Labels needed for training!) ---
# We create 300 random points grouped in 4 blobs (centers=4)
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# --- Step 2: Create Model ---
# We tell it: "Please find 4 groups (clusters)"
kmeans = KMeans(n_clusters=2)
# --- Step 3: Train (Notice: No 'y' here!) ---
kmeans.fit(X)

# --- Step 4: Predictions & Centroids ---
y_kmeans = kmeans.predict(X) # Which group does each point belong to?
centers = kmeans.cluster_centers_ # Where are the centers?

# --- Step 5: Visualize ---
plt.figure(figsize=(8, 6))
# Plot the data points colored by the group K-Means found
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', label='Data Points')

# Plot the Centroids (The "Means")
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.7, marker='X', label='Centroids')

plt.title("K-Means Clustering: Finding Groups Automatically")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# --- Step 1: Same Data (4 Blobs) ---
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# --- Step 2: Loop to calculate Inertia ---
inertia_list = []

for i in range(1, 11): # Test K from 1 to 10
    kmeans = KMeans(n_clusters=i, n_init=10, random_state=42)
    kmeans.fit(X)
    # Inertia value store karo
    inertia_list.append(kmeans.inertia_)

# --- Step 3: Visualize the Elbow ---
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia_list, marker='o', linestyle='--', color='blue')
plt.title('Elbow Method for K-Means')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Distance from Center)')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```
