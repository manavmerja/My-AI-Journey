```python
import pandas as pd

# 1. Setup DataFrame
data = {
    "Student": ["Manav", "Sahil", "Riya"],
    "Marks": [85, 90, 78],
    "Subject": ["Maths", "Science", "Maths"]
}
df = pd.DataFrame(data)
df.set_index("Student", inplace=True) # Student ko index banaya

# ---------------------------------------------------------
# Task 1: loc Slicing (Label based)
# ---------------------------------------------------------
print("Task 1 (loc):")
print(df.loc["Manav":"Sahil"])
# Output mein 'Manav' aur 'Sahil' DONO aayenge.

# ---------------------------------------------------------
# Task 2: iloc Slicing (Position based)
# ---------------------------------------------------------
print("\nTask 2 (iloc):")
print(df.iloc[0:2])
# Output mein Index 0 (Manav) aur Index 1 (Sahil) aayenge.
# Index 2 (Riya) count NAHI hoga.
```

```python
# Sahi Column Name (Subject) aur Sahi Brackets ( (A) | (B) )
smart_science_student = df[(df["Subject"] == "Science") | (df["Marks"] == 90)]

print(smart_science_student)
```

```python
import pandas as pd
import numpy as np

# 1. Ganda Data (Dirty Data)
data = {
    "Student": ["Manav", "Sahil", "Riya", "Amit"],
    "Marks": [85, np.nan, 78, np.nan],  # Sahil aur Amit ke marks gayab
    "Subject": ["Maths", "Science", "Maths", "Maths"]
}
df = pd.DataFrame(data)

# 2. Mean (Average) calculate karo
avg_marks = df["Marks"].mean()
print(f"Average Marks: {avg_marks}")

# 3. Magic: Sirf 'Marks' column ke NaN ko Mean se bhar do
df["Marks"] = df["Marks"].fillna(avg_marks)

print("\nCleaned Data:")
print(df)


# Topper kaun hai? Marks ko 'Descending' (High to Low) order me sort karo
df_sorted = df.sort_values(by="Marks", ascending=False)
print(df_sorted)

# 'Marks' column ka naam badal kar 'Final Score' kar dete hain
df.rename(columns={"Subject": "Course"}, inplace=True)
print(df)
```

```python
import pandas as pd

data = {
    "Student": ["Manav", "Sahil", "Riya", "Amit", "Priya"],
    "Subject": ["Maths", "Science", "Maths", "Science", "Maths"],
    "Marks": [85, 90, 78, 80, 95]
}
df = pd.DataFrame(data)

# Q: Kis Subject mein students ke marks zyada aaye hain? (Subject-wise Average)
subject_avg = df.groupby("Subject")["Marks"].mean()
print(subject_avg)
```

```python
import pandas as pd

# 1. Data Setup (Clean Data Analysis ke liye)
data = {
    "Student": ["Manav", "Sahil", "Riya", "Amit", "Priya"],
    "Subject": ["Maths", "Science", "Maths", "Science", "Maths"],
    "Marks": [85, 90, 78, 80, 95]
}
df = pd.DataFrame(data)

print("Original Data:")
print(df)
print("\n" + "="*30 + "\n")

# ---------------------------------------------------------
# Task 1: Count (Kitne students hain?)
# ---------------------------------------------------------
# Hum 'Subject' ke hisaab se group banayenge aur 'Student' ko ginenge (count)
subject_counts = df.groupby("Subject")["Student"].count()
print("1. Student Count per Subject:")
print(subject_counts)

print("\n" + "-"*20 + "\n")

# ---------------------------------------------------------
# Task 2: Max (Topper marks)
# ---------------------------------------------------------
# Hum 'Marks' column ka max nikalenge
subject_toppers = df.groupby("Subject")["Marks"].max()
print("2. Max Marks per Subject:")
print(subject_toppers)
```

```python
import pandas as pd

data = {
    "Student": ["MANAV", "sahil", "Riya", "MANAV"], # Manav 2 baar hai (Duplicate)
    "Marks": [85, 90, 78, 85]
}
df = pd.DataFrame(data)

# 1. String Cleaning: Sabke naam Title Case (Pehla bada, baaki chota) karte hain
df["Student"] = df["Student"].str.title()
print("1. Name Fixed:\n", df)

# 2. Remove Duplicates: Manav ki copy hatate hain
df = df.drop_duplicates()
print("\n2. Duplicates Removed:\n", df)

# 3. Apply: Custom Logic (Pass/Fail column banana)
def check_result(marks):
    if marks > 80:
        return "Distinction"
    else:
        return "Good"

df["Result"] = df["Marks"].apply(check_result)
print("\n3. Custom Logic Applied:\n", df)
```

```python
import pandas as pd

# --- Step 1: Do Alag Tables Banate hain ---
# Table A: Student ki Info
students = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Manav", "Sahil", "Riya"]
})

# Table B: Student ke Marks
scores = pd.DataFrame({
    "ID": [1, 2, 3],
    "Maths": [85, 90, 78],
    "Science": [88, 92, 80]
})

print("Table A (Students):\n", students)
print("\nTable B (Scores):\n", scores)

# --- Step 2: MERGE (Jodna) ---
# ID ke base par dono ko jod kar ek badi table banayi
merged_df = pd.merge(students, scores, on="ID")
print("\n1. Merged Table (Judi hui):\n", merged_df)

# --- Step 3: MELT (Pighlana/Reshape) ---
# Maths aur Science columns ko tod kar rows bana diya (Visualization ke liye)
melted_df = merged_df.melt(id_vars=["Name"], value_vars=["Maths", "Science"], 
                           var_name="Subject", value_name="Marks")

print("\n2. Melted Table (Long Format for Graphs):\n", melted_df)
```
